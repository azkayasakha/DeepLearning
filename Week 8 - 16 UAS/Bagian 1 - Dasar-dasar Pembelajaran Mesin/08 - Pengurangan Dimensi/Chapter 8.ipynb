📉 1. PCA (Principal Component Analysis)

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Dataset Iris
iris = load_iris()
X = iris.data
y = iris.target

# Proyeksikan ke 2 dimensi dengan PCA
pca = PCA(n_components=2)
X2D = pca.fit_transform(X)
```

🎨 2. Visualisasi hasil PCA

```python
# Plot 2D hasil reduksi dimensi
plt.figure(figsize=(8, 6))
plt.scatter(X2D[:, 0], X2D[:, 1], c=y, cmap="viridis", edgecolor='k')
plt.xlabel("Komponen 1")
plt.ylabel("Komponen 2")
plt.title("PCA - Iris Data (2D Projection)")
plt.colorbar(label="Class")
plt.show()
```

📊 3. Menentukan jumlah komponen optimal (mis. untuk 95% varian)

```python
import numpy as np

pca_full = PCA()
pca_full.fit(X)

cumsum = np.cumsum(pca_full.explained_variance_ratio_)
d = np.argmax(cumsum >= 0.95) + 1  # Komponen untuk menjelaskan ≥ 95% varian

print("Jumlah komponen untuk menjelaskan 95% variansi:", d)
```

⚙️ 4. Kernel PCA (non-linear)

```python
from sklearn.decomposition import KernelPCA

# Uji dengan kernel RBF
kpca = KernelPCA(n_components=2, kernel="rbf", gamma=0.04)
X_kpca = kpca.fit_transform(X)

# Plot
plt.figure(figsize=(8, 6))
plt.scatter(X_kpca[:, 0], X_kpca[:, 1], c=y, cmap="viridis", edgecolor='k')
plt.title("Kernel PCA (RBF) - Iris Data")
plt.xlabel("Komponen 1")
plt.ylabel("Komponen 2")
plt.show()
```

🧩 5. LLE (Locally Linear Embedding)

```python
from sklearn.manifold import LocallyLinearEmbedding

lle = LocallyLinearEmbedding(n_components=2, n_neighbors=10)
X_lle = lle.fit_transform(X)

plt.figure(figsize=(8, 6))
plt.scatter(X_lle[:, 0], X_lle[:, 1], c=y, cmap="viridis", edgecolor='k')
plt.title("LLE - Iris Data")
plt.xlabel("Dimensi 1")
plt.ylabel("Dimensi 2")
plt.show()
```

## ✅ Penjelasan:

* **PCA** cocok untuk data linear → cepat dan efisien.
* **Kernel PCA** digunakan saat data **tidak dapat dipisahkan secara linear**.
* **LLE** cocok untuk **manifold learning**, menjaga struktur lokal data.
